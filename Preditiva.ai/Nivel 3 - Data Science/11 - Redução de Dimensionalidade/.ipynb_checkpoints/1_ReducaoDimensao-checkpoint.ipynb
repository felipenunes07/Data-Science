# Redução de Dimensionalidade: PCA

## Importação das bibliotecas

```python
# Pacotes de preparação dos dados
import numpy as np
import pandas as pd

# Pacotes gráficos
import matplotlib.pyplot as plt
import seaborn as sns

# Pacotes de modelagem
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

## Geração de Dados Sintéticos

```python
np.random.seed(42)
x1 = np.linspace(1, 10, 10)
x2 = 1 + x1 * 2 + np.random.normal(0.2, 1, 10)
```

```python
df = pd.DataFrame(data=np.array([x1,x2]).T,
                  index=None,
                  columns=['x1','x2'])
df
```

```python
plt.plot(df['x1'], df['x2'], 'o');
```

## Padronização da Escala

```python
# Padronização da Escala
x_scaler = StandardScaler()
x_scaler.fit(df)
x_scaled = x_scaler.transform(df)
```

```python
# Média de X1 e X2
x_scaled.mean(axis=0)
```

```python
# Desvio Padrão de X1 e X2
x_scaled.std(axis=0)
```

```python
plt.plot(x_scaled[:,0],
         x_scaled[:,1],
         'o');
```

## Aplicação do PCA
Utilizaremos a técnica PCA disponível na biblioteca Scikit-Learn.

```python
pca = PCA(n_components=2,
          random_state=42)
z = pca.fit_transform(x_scaled)
```

```python
z
```

### Qual o percentual de Retenção de Variabilidade?

```python
pca.explained_variance_ratio_
```

### Usando o valor de Z para tentar voltar as dados originais

```python
# Voltando para os dados originais
X_aproximado = pca.inverse_transform(z)
```

```python
# Dados Originais
plt.plot(x_scaled[:,0],
         x_scaled[:,1],
         'o', label="Dados Originais")

# Dados Aproximados
plt.plot(X_aproximado[:,0],
         X_aproximado[:,1],
         'o', color="red", label="Dados Aproximados")
plt.legend()

plt.show();
```

# Redução de Dimensionalidade: AutoEncoder

## Importação das bibliotecas

```python
# Pacotes de modelagem
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
```

## Aplicação do AutoEncoder

```python
x_scaled.shape[1]
```

```python
# Definição da camada de entrada
input_data = layers.Input(shape=(x_scaled.shape[1],))

# Definição das camadas de codificação
encoded1 = layers.Dense(128, activation='tanh')(input_data)
encoded2 = layers.Dense(64, activation='tanh')(encoded1)
encoded3 = layers.Dense(32, activation='tanh')(encoded2)
encoded4 = layers.Dense(16, activation='tanh')(encoded3)
encoded5 = layers.Dense(8, activation='tanh')(encoded4)
encoded6 = layers.Dense(4, activation='tanh')(encoded5)
encoded7 = layers.Dense(2, activation='tanh')(encoded6)

# Definição das camadas de decodificação
decoded1 = layers.Dense(4, activation='tanh')(encoded7)
decoded2 = layers.Dense(8, activation='tanh')(decoded1)
decoded3 = layers.Dense(16, activation='tanh')(decoded2)
decoded4 = layers.Dense(32, activation='tanh')(decoded3)
decoded5 = layers.Dense(64, activation='tanh')(decoded4)
decoded6 = layers.Dense(128, activation='tanh')(decoded5)
decoded7 = layers.Dense(x_scaled.shape[1], activation='linear')(decoded6)

# Treinamento do AutoEncoder
autoencoder = keras.Model(input_data, decoded7)
autoencoder.compile(optimizer='adam', 
                    loss='mean_squared_error')

autoencoder.summary()
```

```python
(4*128)+128
```

```python
autoencoder.fit(x_scaled, x_scaled,
                epochs=100,
                batch_size=16,
                shuffle=True)
```

## Extração dos Dados Comprimidos

```python
# Encoder
encoder = keras.Model(input_data, encoded7)

# Decoder
decoder = keras.Model(input_data, decoded7)

# Transformação dos dados
encoded_data = encoder.predict(x_scaled)
decoded_data = decoder.predict(x_scaled)
```

```python
encoded_data
```

```python
# Gráfico de dispersão das dimensões do AutoEncoder
palette = sns.color_palette("bright", 3)
sns.scatterplot(encoded_data[:,0], 
                encoded_data[:,1], 
                hue=df['Species'], 
                legend='full', 
                palette=palette);
```

```python
decoded_data
```

```python
# Gráfico de dispersão das dimensões do AutoEncoder
palette = sns.color_palette("bright", 3)
sns.scatterplot(decoded_data[:,1], 
                decoded_data[:,3], 
                hue=df['Species'], 
                legend='full', 
                palette=palette);
```
